```python
from PIL import Image, ImageDraw, ImageFilter
#import matplotlib.pyplot as plt

ims = Image.open('kid_1.jpg')

blur_radius = 5
offset = 1
back_color = Image.new(ims.mode, ims.size, (220,20,60))
offset = blur_radius * 2 + offset
mask = Image.new("L", ims.size, 0)
draw = ImageDraw.Draw(mask)

draw.rounded_rectangle((offset, offset, ims.size[0] - offset, ims.size[1] - offset), fill=255)

mask = mask.filter(ImageFilter.GaussianBlur(blur_radius))

ims_round = Image.composite(ims, back_color, mask)
# plt.imshow(ims_round)
# ims_round.save('./kid_1_mask.jpg', quality=95)
ims_round
```
