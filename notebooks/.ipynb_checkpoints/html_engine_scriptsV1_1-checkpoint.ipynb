```python
for page in full_book_json['templates']:
    """What ever on book level for font should be available on page level if not defined there..."""
    font_attributes_for_deligation = ['font_color', 'font_size', 'font_family', 'font_style']
    for attribute in font_attributes_for_deligation:    
        if attribute not in page:
            page[attribute] = full_book_json[attribute]            
    """Now what on the page level for the font put it on type:text element level"""    
    for attribute in font_attributes_for_deligation:
        for element in page['elements']:
            for character in element['character']:
                if attribute not in character and character['type'] in ['text']:
                    character[attribute] = page[attribute]            
                """Now what is on element level put it on the line level"""
                if character['type'] in ['text']:
                    genders = ['boy', 'girl']
                    elemenent_style = character['style']
                    for gender in genders:                        
                        for line in character['script'][gender]:
                            if attribute not in line:
                                line[attribute] = character[attribute]    
    """Now we want to override or extend the style on the element level to line level
    for elements of type:text"""
    #---possible valuesa are extend and override
    style_behaviour = 'extend'
    for element in page['elements']:
        for character in element['character']:
            if character['type'] in ['text']:
                text_element_style = character['style']
                genders = ['boy', 'girl']
                for gender in genders:   
                    for line in character['script'][gender]:
                        if style_behaviour in ['override']:
                            line['style'] = text_element_style
                        else:
                            line['style'].extend(text_element_style)
```
