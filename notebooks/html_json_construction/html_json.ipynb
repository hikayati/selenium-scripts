```python
import json
script = None
with open("./english.json", "r") as f:
    script = json.loads( f.read() )
```

```python
import copy
import string

global_group_index = 0

groups_single_letter = script['groups'][0]

all_groups = []

def replace_place_holders(group, variation, letter=''):
    
    """If we have less then four templates then copy first to others"""
    if len(group['templates']) < 3:        
        for i in range(len(group['templates']), 4):            
            new_template = copy.deepcopy(group['templates'][0])                
            new_template['template_index'] = str(i)    
            group['templates'].append(new_template)    
    
    for index, template in enumerate(group['templates']):
        print("Template index:", index)
        group_index = group['group_index']
        template['template_index'] = str(index)
        template['template_tag'] = group['group_tag']
        #text = template['elements'][0]['character'][1]['script']['boy'][0]['text']
        #print("[Before ]text:", text)
        #text = text.replace("{letter}", letter).replace("{group_index}", '').replace("{template_index}", template_index).replace("--", "-")                
        text = f"{group['group_tag']}-{index}"
        print("[After ]text:", text)        
        template['elements'][0]['character'][1]['script']['boy'][0]['text'] = text        
        template['elements'][0]['character'][1]['script']['girl'][0]['text'] = text              
        template['elements'][0]['character'][1]['script']['boy'][0]['style'] = [ "position:absolute;", "right:100px;", "font-size: 100px;"]       
        template['elements'][0]['character'][1]['script']['girl'][0]['style'] = [ "position:absolute;", "right:100px;","font-size: 100px;"]                        
        group['templates'][index] = template
        
        """Delete the second character..."""
        try:
            del template['elements'][1]
        except:
            pass        
    return group

for group_index, letter in enumerate(string.ascii_uppercase):    
    for variation in range(0, 2):
        print("--"*20)        
        group = copy.deepcopy(groups_single_letter)    
        print("Group Index:", global_group_index)
        print("Letter:", letter)
        group['group_tag'] = group['group_tag'].replace("{letter}", letter)    
        group['group_index'] = str(global_group_index)
        group = replace_place_holders(group, variation, letter)    
        all_groups.append(group)    
        global_group_index += 1
    
other_groups = script['groups'][1:len(script['groups'])]
    
for index, og in enumerate(other_groups):    
    for variation in range(0, 2):
        print("**"*20)
        group = copy.deepcopy(og)    
        print("Group Index:", global_group_index)
        group['group_index'] = str(global_group_index)
        group = replace_place_holders(group, variation, letter)
        all_groups.append(group)
        global_group_index += 1
```

```python
script = None
with open("./english.json", "r") as f:
    script = json.loads( f.read() )
script['groups'] = all_groups
```

```python
with open("./english_new.json", "w") as f:
    f.write( json.dumps(script, indent=3) )
```

```python
"""Need an algorithm where we will provide the name and it will give us the tags, """
sub_story_order = ['Intro', ""]

#"sha" : [ "Intro", "1 Letter", "StoryFilling", "StoryFilling", "2 Letter", "3 Letter", "StoryEnding" ]
#"shah" : [ "Intro", "1 Letter", "StoryFilling", "2 Letter", "3 Letter", "4 Letter","StoryEnding" ]
#"shahhh" : ["Intro", "1"]
```

```python
def get_group_tag_order(name):
    group_tag_freq = {
        "StoryFilling" : 2,
        "Letter" : 2,
        "StoryEnding" : 2,
        "GeneralLetter" : 2,
        "Intro" : 2
    }    
    if len(name) == 3:
        sub_story_order = [ "Intro", "0 Alphabet", "StoryFilling", "StoryFilling", "1 Alphabet", "2 Alphabet", "StoryEnding" ]
    elif len(name)==4:
        sub_story_order = [ "Intro", "0 Alphabet", "StoryFilling", "1 Alphabet", "2 Alphabet", "3 Alphabet","StoryEnding" ]
    else:
        sub_story_order = ['Intro']
        letter_frequency = {}
        for index, letter in enumerate(name):
            if letter not in letter_frequency:
                letter_frequency[letter] = 1
            else:
                letter_frequency[letter] += 1

            if letter_frequency[letter] <=  group_tag_freq["Letter"]:
                sub_story_order.append( str(index) + " Alphabet")
            else:
                sub_story_order.append("DynamicLetter-"+letter)
        sub_story_order.append("StoryEnding")        
    #----here
    #print("sub_story_order", sub_story_order)
    for tags_index, group_tag in enumerate(sub_story_order):
        if "Alphabet" not in group_tag:
            continue
        index = int(group_tag.replace(" Alphabet", ""))
        letter = name[index]
        sub_story_order[tags_index] = "Letter" + letter.upper()
        
    return sub_story_order
```

```python
def get_templates_of_selected_groups(sub_story_order, book_json):
    visited_group_index = []
    selected_groups = []
    selected_group_tags = []
    for group_tag in sub_story_order:
        print("Finding Group Tag:", group_tag)
        for group in book_json['groups']:
            if group['group_tag'] in group_tag and group['group_index'] not in visited_group_index:
                print("Appending Group with Tag:", group['group_tag'])                
                
                if "letter" in group:
                    group['letter'] = group_tag.split("-")[-1]
                
                selected_groups.append(group)
                visited_group_index.append(group['group_index'])
                selected_group_tags.append(group_tag)
                break
    all_templates = []
    for selected_group in selected_groups:
        all_templates.extend( selected_group['templates'] )
    return selected_groups
```

```python
### Making arabic

import json
script = None
with open("./english.json", "r") as f:
    script = json.loads( f.read() )
```

```python
for group in script['groups']:
    for template_no, template in enumerate(group['templates']):
#         if element['type'] == 'avatar':            
#             del template['elements'][0]['character'][0]    
        template['elements'] = [ template['elements'][0] ]
        for element in template['elements'][0]['character']:                
            
            if element['type'] == 'avatar':
                
                element['image_name'] = '/gallery'            
            
            if element['type'] == 'text':
                """Now we need some changes..."""
                text_english = {                    
                    "boy" : [
                        {
                            "text" : f"English Boy {group['group_tag']}-{template_no}",
                            "style" : ["font-size:100px"],
                            "tag_name" : "p"
                        }
                    ],
                    "girl" : [
                        {
                            "text" : f"English Girl {group['group_tag']}-{template_no}",
                            "style" : ["font-size:100px"],
                            "tag_name" : "p"
                        }
                    ]
                }
                text_arabic = {                    
                    "boy" : [
                        {
                            "text" : f"الفتى العربي {group['group_tag']}-{template_no}",
                            "style" : ["font-size:100px"],
                            "tag_name" : "p"
                        }
                    ],
                    "girl" : [
                        {
                            "text" : f"فتاة عربية {group['group_tag']}-{template_no}",
                            "style" : ["font-size:100px"],
                            "tag_name" : "p"
                        }
                    ]
                }
                element['script_english'] = text_english
                element['script_arabic'] = text_arabic
                del element['script']
```
